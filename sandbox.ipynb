```python
import numpy as np
from pystream import streamgraph
import matplotlib.pyplot as plt

data = [np.random.randn(100).cumsum() for _ in range(5)]
labels = ["Stream 1", "Stream 2", "Stream 3", "Stream 4", "Stream 5"]
colors = ['b', 'g', 'r', 'c', 'm']

fig, ax = streamgraph(data, labels, colors, smoothing_method='interpolation', window_size=5)
plt.show()
```
